# 🔧 Run Configuration
Set your experiment’s metadata and parameters here. Update before every run.

```python
dataset_version = "3"
change_number = "4"
model_id = "M06"
model_used = "yolov8s.pt"
epochs = 60
image_size = 960
augmentations = "increased image size to 960, and batch size to 64"
classes = "Shark"  # or "All"
null_images = 1462
notes = "increased image size to 960, and batch size to 64"
```

# 🚀 Train the Model
Train your YOLOv8 model using the settings above.

```python
model = YOLO(model_used)
results = model.train(data='/content/F.I.N.S-one-class-3/data.yaml', epochs=epochs, imgsz=image_size, batch=64)
```

# 📊 Extract and Save Metrics
Automatically log results like mAP, precision, and recall to a tracking CSV.

```python
# Access the training arguments from the model object
args = model.args
save_dir = results.save_dir
folder_name = os.path.basename(save_dir)

csv_path = "/content/drive/MyDrive/F.I.N.S_IMPLEMENTATION_FOLDER/F.I.N.S_PERFORMANCE_TRACKING_CSV.csv"

metrics = {
    "mAP50": results.box.map50,
    "mAP50-95": results.box.map,
    "precision": results.box.p.item(),
    "recall": results.box.r.item()
}


import pytz
from datetime import datetime

local_tz = pytz.timezone("Australia/Brisbane")
now = datetime.now(local_tz)
timestamp = now.strftime("%d%b_%H-%M")

# CSV Columns:
# ["Date/Time", "Dataset Version", "Change #", "Model ID", "Model Used", "Epochs",
#  "mAP@0.5", "mAP@0.5–0.95", "Precision", "Recall", "Augmentations",
#  "Image Size", "Classes", "Null Images", "Run Folder", "Notes"]

row = [
    timestamp,
    dataset_version,
    change_number,
    model_id,
    model_used,
    epochs,
    metrics["mAP50"],
    metrics["mAP50-95"],
    metrics["precision"],
    metrics["recall"],
    augmentations,
    f"{image_size}x{image_size}",
    classes,
    null_images,
    folder_name,
    notes
]

with open(csv_path, "a", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(row)
```

# 💽 Backup Model Folder
Save the full YOLO training folder (weights, graphs, logs) to Drive.

```python

named_folder = f"{model_id}_{timestamp}_{folder_name}"
backup_path = f"/content/drive/MyDrive/yolo_models/{named_folder}"
shutil.copytree(save_dir, backup_path)
print(f" Model folder backed up to: {backup_path}")
```

Upload a test image and run prediction using your trained model.

```python
uploaded = files.upload()
refined_model = YOLO(f"{save_dir}/weights/best.pt")

for filename in uploaded:
    results = refined_model.predict(source=filename, save=True, show=False, conf=0.1)
    predict_dirs = sorted(glob.glob("runs/detect/predict*"), key=os.path.getmtime)
    latest_dir = predict_dirs[-1]
    print(f"✅ Prediction saved in: {latest_dir}")
```
